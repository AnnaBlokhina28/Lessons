```python
summ = 0
minus_var = ""
n=input("  :").strip().replace(".","").replace(",", "").replace(" ", "")

if "-" in n:
    minus_var = "-"

n = n.replace("-", "") 
    
for i in range(len(n)):
    summ += int(n[i])

print(minus_var + str(summ))
```

```python
N = int(input(" : ")) 
p=[]

for i in range(0, N): 
    if i == 0:
        p.append(1)
    else: 
        p.append(p[i-1]*(i+1))
print(p)
```
