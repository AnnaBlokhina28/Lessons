### 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/esratmaria/house-price-dataset-with-other-information

### 1.2 Изучите стоимости недвижимости

Вывод: Большая часть недвижимости имеет стоимость до 900 тыс. долларов. Меньше всего недвижимости в данных имеет стоимость от 2 мил. долларов и выше

```python
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Кол-во')
```

<font color="#C71585"><h2><b>Задача 2</b></h2></font>

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

Вывод: Близость к набережной имеет маленькое колличество домов всего 0.8 процентов от всего количества.

```python
# Готовим данные для графика
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
    
plt.figure(figsize=(6, 4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Доля домов близких к набережной')
plt.legend(data.index)
```

### 2.2 Изучите распределение этажей домов

Вывод: Большее количество домов имеет один этаж. Меньше домов с этажами больше 2 и ещё меньше с 3 этажами. Наблюдается обратная линейная зависимость этажности домов от их количества, если не считать дома с не типичной этажностью, имеющих чердак, мансарду или подвал.

```python
counts = df['floors'].value_counts()

plt.figure(figsize=(8, 6))    
plt.title('Распредение этажей')
sns.barplot(x=counts.index, y=counts.values)
    
plt.show()
```

### 2.2 Изучите распределение состояния домов

Вывод: Больше всего домов имеют удовлетворительное состояния - оценку 3. Меньше всего домов имеют плохую оценку состояния дома-1.

```python
counts = df['condition'].value_counts()

plt.figure(figsize=(8, 6))    
plt.title('Распредение состояния домов')
sns.barplot(x=counts.index, y=counts.values)
    
plt.show()
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

<br><b>Вывод:</b> Если вы собираетесь купить дом и боитесь прогадать в цене! Хотите с красивым видом, максимальной площади и в хорошем состоянии и за приемлемые деньги? Сейчас мы разберем от чего зависти цена на недвижимость на примере данных из 21613 объектов, находящихся северо-западе США в штате Вашингтон.</br> 
<br>Цена домов на рынке варьируется от 75000 до 7700000 долларов. Для изучения факторов, влияющих на стоимость недвижимости, была использована таблица корреляционного анализа. Было обнаружено, что несколько переменных играют решающую роль в определении цены дома. Первый фактор - это жилая площадь дома.</br>
<br>Чем больше жилая площадь, тем выше, вероятно, будет цена дома. Второй фактор - это количество ванных комнат в доме. Большее количество ванных комнат в доме обычно приводит к более высокой цене.</br>
<br>Другим фактором, влияющим на стоимость дома, является оценка недвижимости или сорт. Дома с более высоким классом, как правило, стоят дороже. Наконец, наличие чердака или подвала также влияет на цену дома. Дома с этими функциями, как правило, дороже, чем дома без них.</br>
<br>Одно интригующее открытие заключается в том, что двухэтажные дома, как правило, стоят дороже несмотря на то, что они встречаются реже, чем одноэтажные дома. Данные свидетельствуют о том, что это может быть связано с более высоким спросом на двухэтажные дома, которые часто воспринимаются как более просторные и роскошные. А вот на дома с тремя этажами будет уже немного ниже, как не странно.</br> 
<br>Состояние дома также влияет на его цену. Корреляционный анализ показал, что существует почти равномерная зависимость от состояния дома. Однако дома с плохой оценкой состояния продаются немного дешевле, как видно из графика.</br>
<br>Прекрасный вид из окна значительно влияет на стоимость дома наблюдается прямая линейная зависимость на графике, но в данных очень много домов с оценкой вида - 0, из-за большего количества можно сказать, что цена на многие дома высока, хоть они и не выходят на красивый вид. Наконец, близость дома к набережной также влияет на цену дома. Была обнаружена небольшая прямая линейная зависимость между стоимостью дома и близостью к набережной. Те немногие дома, которые находятся близко к набережной, как правило, стоят дороже. Однако количество домов рядом с набережной невелико, поэтому корреляция незначительна.</br>
