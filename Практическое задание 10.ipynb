```python
Задача: написать телеграмм-бота для игры в Камень-Ножницы-Бумага
После запуска (команда /start ) бот должен предложить нажать кнопки и начать игру.
Добавить три кнопки.
После нажатия любой кнопки бот отвечает Камень, Ножницы или Бумага.
Домашнее задание: Добавить обработку ходов и подсчет очков.
```

```python
import telebot
import random
from telebot import types
import tk1

bot = telebot.TeleBot(tk1.Token)
game_bot = 0
game_me = 0

users = {}

def new_user(message):
    global users
    
    chatid = message.chat.id
    
    if not chatid in users:
        users[chatid] = {}
        users[chatid]["game_me"] = 0
        users[chatid]["game_bot"] = 0
        users[chatid]["user"] = message.chat.username or "No Name"

@bot.message_handler(commands=["start"])
def start(message, res=False):
  # Добавляем кнопки
    markup=types.ReplyKeyboardMarkup(resize_keyboard=True)
    item1=types.KeyboardButton("Камень")
    item2=types.KeyboardButton("Ножницы")
    item3=types.KeyboardButton("Бумага")
    # TODO добавить 3 объекта KeyboardButton для 'Камень','Ножницы','Бумага'
    markup.add(item1)
    markup.add(item2)
    markup.add(item3)                           
    # TODO добавить кнопки в клавиатуру (markup)

    bot.send_message(message.chat.id, 'Нажми кнопку и начни игру ',  reply_markup=markup)
    
@bot.message_handler(commands=["del"])
def start_del(message, res=False):
    
    if "del" in message.text:
        global game_me
        global game_bot
        
        new_user(message)
        
        users[message.chat.id]["game_bot"] = 0
        users[message.chat.id]["game_me"] = 0
        
        bot.send_message(message.chat.id,"Результаты игры очищены!")

        
@bot.message_handler(commands=["rank"])
def start_rank(message, res=False):
    if "rank" in message.text:
        result = "Рейтинг игроков: \n\n"
        for item in users:
            result += str(users[item]["user"]) + ": \n" + "Игрок: " + str(users[item]["game_me"]) + "\n" + "Бот: "   + str(users[item]["game_bot"]) + "\n" + ("-"*30) + "\n"
        
        bot.send_message(message.chat.id, result)
        
@bot.message_handler(content_types=["text"])
def handle_text(message):
    #bot.send_message(message.chat.id,message.text)
    # TODO посылать в чат случайное из списка 'Камень','Ножницы','Бумага'
    game = ("Камень", "Ножницы", "Бумага")
    if message.text in game:
        answer = random.choice(game)
        bot.send_message(message.chat.id, answer)
        global game_me
        global game_bot
        global users
        
        new_user(message)
        
        if message.text == "Ножницы":
            if answer == "Ножницы":
                result = "&#129335; <b>Ничья</b>"
                #game_me +=1
                #game_bot +=1
            elif answer == "Бумага":
                result = "&#127942; <b>Вы выиграли</b>"
                game_me +=1
                users[message.chat.id]["game_me"] +=1
            else:
                result = "&#9940; <b>Вы проиграли</b>"
                game_bot +=1
                users[message.chat.id]["game_bot"] +=1

        if message.text == "Бумага":
            if answer == "Бумага":
                result = "&#129335; <b>Ничья</b>"
                #game_me +=1
                #game_bot +=1
            elif answer == "Камень":
                result = "&#127942; <b>Вы выиграли</b>"
                game_me +=1
                users[message.chat.id]["game_me"] +=1
            else:
                result = "&#9940; <b>Вы проиграли</b>"
                game_bot +=1
                users[message.chat.id]["game_bot"] +=1

        if message.text == "Камень":
            if answer == "Камень":
                result = "&#129335; <b>Ничья</b>"
                #game_me +=1
                #game_bot +=1
            elif answer == "Ножницы":
                result = "&#127942; <b>Вы выиграли</b>"
                game_me +=1
                users[message.chat.id]["game_me"] +=1
            else:
                result = "&#9940; <b>Вы проиграли</b>"
                game_bot +=1
                users[message.chat.id]["game_bot"] +=1

        bot.send_message(message.chat.id, f'{answer}\n\n{result}\n\nОбщий счёт \nbot - {users[message.chat.id]["game_bot"]} \nВы - {users[message.chat.id]["game_me"]}', parse_mode='HTML')

bot.polling(none_stop=True, interval=0)
```
