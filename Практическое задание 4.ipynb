```python
import math

num = int(input())
list_simple = []
simple = 2
while num > 1:
    if num % simple == 0:
        list_simple.append(simple)
        num = num/simple
    else:
        simple += 1
if len(list_simple)<=1:
    print(list_simple[0]," -  ")
else:
    print(math.prod(list_simple),"  :", list_simple)
    
```
