<font color="#56a7cf"><h2><b>Формулы</b></h2></font>

```python
import math
from math import factorial as f
import pandas as pd

def C(n, k):
    return f(n) / (f(k) * f(n-k))

def Bernoulli(n, k, p):
    q=1-p
    return C(n, k)*(p**k)*(q**(n-k))

def Poisson(m,lmd):
    return (lmd**m/math.factorial(m))*math.e**-lmd
```

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
#1.1 Найдем среднее арифметическое: 

Salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
Srarif = sum(Salary)/len(Salary)
print(f"Среднее арифметическое = {Srarif}")
```

```python
#1.2 Смещанная и несмещенная дисперсии для данной выборки:

Salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
Srarif = 65.3
SummaX = []

for i in Salary:
    SummaX.append((i-Srarif)**2)

#Смещенная дисперия
D1 = sum(SummaX)/len(Salary)
print(f"Смещенная дисперсия = {round(D1)}")

#Несмещенная дисперия
D2 = sum(SummaX)/(len(Salary))-1
print(f"Несмещенная дисперсия = {round(D2)}")
```

```python
#1.3 Среднее квадратическое отклонение:

D1 = 950
D2 = 949
SKO1 = math.sqrt(D1)
SKO2 = math.sqrt(D2)
print(f"Смещенное стандартное отклонение = {round(SKO1,3)}")
print(f"Несмещенное стандартное отклонение = {round(SKO2,3)}")
```

<font color="#56a7cf"><h2><b>Задание 2</b></h2></font>

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4.
Какова вероятность того, что 3 мяча белые?

Решение:

Р(A)-взяли 3 мяча белые; Р(A) = m/n, n = C(8,2)*C(12,4),
m -благоприятные сочетания: из 1 ящика 2 бел, из 2-го 1 бел ИЛИ из 1 ящика 1 бел, из 2-го 2 белых ИЛИ из 1ящика 0 белых, из второго 3 белых

m=C(5,2)*C(5,1)*С(3,7)+С(5,1)*С(3,1)*С(5,2)*С(7,2)+С(2,3)*С(5,3)*С(1,7)

```python
n = C(8,2)*C(12,4)
m = C(5,2)*C(5,1)*C(7,3)+C(5,1)*C(3,1)*C(5,2)*C(7,2)+C(3,2)*C(5,3)*C(7,1)
P = m/n
print(f"Вероятность, что 3 мяча белые  = {round(P,2)}")
```

<font color="#56a7cf"><h2><b>Задание 3</b></h2></font>

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Решение:Так как событие уже произошло - попали в мешень, то используем формулу Байеса:
Р(A) - попадение в мешень, P(B) - стрелял один из спортсменов 
P(B) - считаем, что у всех спортсменов были одинаковые шансы стрелять на соревноваяниях т.е вероятность 1/3 для каждого P(B1)=P(B2)=P(B3)=0.333

```python
PAB1 = 0.9
PAB2 = 0.8
PAB3 = 0.6
PB = 1/3

#Найдем полную вероятность:
PA = PAB1*PB+PAB2*PB+PAB3*PB

#3.a Вероятность, что попал первый спортсмен:

PBA1 = (PAB1*PB)/PA
print(f"3.a Вероятность, что попал первый спортсмен  = {round(PBA1,2)}")

#3.б Вероятность, что попал второй спортсмен:

PBA2 = (PAB2*PB)/PA
print(f"3.б Вероятность, что попал второй спортсмен  = {round(PBA2,2)}")

#3.в Вероятность, что попал третий спортсмен:

PBA3 = (PAB3*PB)/PA
print(f"3.в Вероятность, что попал третий спортсмен  = {round(PBA3,2)}")
```

<font color="#56a7cf"><h2><b>Задание 5</b></h2></font>

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

Решение:

5.а Р(А) - вероятность, что все детали выйдут из строя Р(А) = p1*p2*p3
5.б P(B) - вероятность, что только 2 детали выйдут из строя P(B) = p1*p2*q3+p1*q2*p3+q1*p2*p3
5.в P(notC) - вероятность, что вес детали не выйдут из строя Р(notС) = q1*q2*q3 
 P(C)-вероятность, что хоть одна сломается Р(С) = 1-Р(notС)
5.г P(D) - вероятность, что от 1-2 детали выйдут из строя Р(D) = p1*q2*q3+p2*q1*q2+p3*q1*q2+P(B)


```python
p1=0.1
p2=0.2
p3=0.25

#5.а Р(А) - вероятность, что все детали выйдут из строя Р(А) = p1*p2*p3
PA = p1*p2*p3
print(f"5.a Вероятность, все детали выйдут из строя Р(А)  = {round(PA,3)}")

#5.б Р(B) - вероятность, что только 2 детали выйдут из строя P(B) = p1*p2*q3+p1*q2*p3+q1*p2*p3
PB = p1*p2*(1-p3)+p1*(1-p2)*p3+(1-p1)*p2*p3
print(f"5.б Вероятность, что только 2 детали выйдут из строя P(B)  = {round(PB,3)}")

#5.в Р(C) - вероятность, что хоть одна сломается
РnotС = (1-p1)*(1-p2)*(1-p3)
РС = 1-РnotС
print(f"5.в Вероятность, что хоть одна сломается Р(C)  = {round(РС,3)}")

#5.г Р(D) - вероятность, что от 1-2 детали выйдут из строя Р(D) = p1*q2*q3+p2*q1*q2+p3*q1*q2+P(B)
РD = p1*(1-p2)*(1-p3)+p2*(1-p1)*(1-p3)+p3*(1-p1)*(1-p2)+PB
print(f"5.г Вероятность, что от 1-2 детали выйдут из строя Р(D)  = {round(РD,3)}")
```
