```python
def TOFloat(a=""):
    '''
           :
    : 
    "6.012" -> 0.012
    '''
    b = -1

    try:
        b = float("0."+ a.split(".")[1])
    except:
        b = -1

    return b
```

```python
A = input("  (c )   :").split()
b = []
for i in A:
    b.append(TOFloat(i))
print(max(b) - min(b))
```

```python
def fibonacci(n):
    if (n <= 1):
        if (n<=-2):
            n = n*-1
            return ((fibonacci(n-1) + fibonacci(n-2))*-1)
        else:
            if n == -1:
                return 1
            else:
                return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))
    
n = int(input("   :"))
print(" :")
for i in range(n*-1, n+1):
    
    print(fibonacci(i))
```
