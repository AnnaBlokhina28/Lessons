```python
temp=""
s_1=["AD", "BD", "AC","BC"]

s = input("   x  y  :")
s = s.split(",")
print(s)


for i in range(len(s)):
    if s[i].strip() == '0':
        print("   0")
        break
    else:
        if i==0:
            if float(s[i])>0:
                temp+="B"
            else: temp+="A"
        
        if i==1:
            if float(s[i])>0:
                temp+="D"
            else: temp+="C"
                
for i in range(len(s_1)):
    if s_1[i]==temp:
        print(i+1,"")
        break
```
